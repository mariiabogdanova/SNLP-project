```python
"""
This is the code for exaluating average similarity calculated from the multiple similarity scores estimated by human correspondents.
"""

import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv("LeePincombeWelshData.csv")
data.head(10)
```

```python
def all_pairs(lst):
    """
    This function creates all unique value pairs from list of values. Returns tuples.
    """
    return [ (lst[i],lst[j]) for i in range(0,len(lst)) for j in range(i+1,len(lst)) ]

def calculate_average_similarity(docs, data):
    """
    This function calculates average similarity for the given documents tuple in the given data.
    """
    return np.mean(data.loc[data['Document1'].isin([docs[0], docs[1]]) & data['Document2'].isin([docs[0], docs[1]])].Similarity)

def normalize_data(data):
    """
    This function normalizes given data (min = 1, max = 5)
    """
    return (data - 1) / (5 - np.min(data))
```

```python
similarities = []
pair_1 = []
pair_2 = []
for pair in all_pairs(np.arange(1,50)):
    pair_1.append(pair[0])
    pair_2.append(pair[1])
    similarities.append(calculate_average_similarity(pair, data))
```

```python
d = {'Document_1': pair_1, 'Document_2': pair_2, 'Similarity_avg': similarities}
df = pd.DataFrame(data=d)
```

```python
df["Similarity_avg_normalized"] = normalize_data(df.Similarity_avg.values.reshape(-1, 1))
```

```python
df.to_csv('AverageSimilarities.csv', index=False)
df.head(10)
```
